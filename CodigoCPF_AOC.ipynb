```python
cpf = [0,2,1,5,1,9,9,7,6]

# PRIMEIRO DÍGITO VERIFICADOR
# multiplicar cada um de 10 a 2
# dividir o resultado somado por 11
# se o dígito for 0 ou 1 coloca 0
# se for maior q 2, você subtrai 11 pelo valor 


# PRIMEIRO DÍGITO VERIFICADOR
pridig = []
i = 10
for x in cpf:
    a = i * x
    pridig.append(a)
    i -= 1

soma1 = 0
for x in pridig:
    soma1 +=x


n = soma1/11

if (n == 0 or n == 1):
    pd = 0
else:
    pd = 11-n

#SEGUNDO DÍGITO VERIFICADOR
# multiplicar cada um de 11 a 3
# dividir o resultado somado por 11
# se o dígito for 0 ou 1 coloca 0
# se for maior q 2, você subtrai 11 pelo valor 

#SEGUNDO DÍGITO VERIFICADOR
segdig = []
j = 10
for x in cpf:
    b = j * x
    segdig.append(b)
    j -= 1

soma2 = 0
for x in segdig:
    soma2 += x

y = soma2/11

if (y == 0 or y == 1):
    sd = 0
else:
    sd = 11-y

print(f"Esse é o primeiro número: {pd}")
print(f"Esse é o segundo número: {sd}")
```
